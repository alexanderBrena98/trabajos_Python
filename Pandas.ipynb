Objetivos:
Cargar un conjunto de datos de estudiantes universitarios.
Realizar análisis estadístico básico.

### Pasos:

1.Cargar los datos: Utilizar Pandas para cargar un archivo CSV que contiene los datos de los estudiantes

```python
import pandas as pd
dataframe = pd.read_csv("/content/drive/MyDrive/Data Science Research Python/universidad_data.csv",sep=",",header=0)
dataframe
```

* 2.Inspeccionar los datos: Mostrar las primeras 5 filas del DataFrame para tener una idea general de los datos.

```python
df = dataframe.head(5)
df
```

* 3.Estadísticas básicas: Utilizar el método describe para obtener estadísticas resumidas de las columnas numéricas.

```python
df = dataframe.head(100)
df.drop('ID del Estudiante', axis=1, inplace=True)
df.describe().round()
```

* 4.Número de estudiantes por Universidad: Utilizar groupby para contar el número de estudiantes de cada Universidad.

```python
df = dataframe.head(100)
df.groupby('Universidad')['Universidad'].count()
```

* 5.Histograma de Edades: Utilizar Matplotlib para crear un histograma que muestre la distribución de las edades de los estudiantes.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.hist(df['Edad'], color='red')
plt.title('Histograma de Edades')
plt.xlabel('Edades')
plt.ylabel('Numero Estudiantes')
plt.show()
```
